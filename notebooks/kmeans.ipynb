##  Data Preparation and Cleaning

```python
import pandas as pd

# Load data
train_df = pd.read_csv('../data/train_2.csv')
```

```python
# Assuming train_df is already loaded
missing_data = train_df.isnull().sum()
missing_data
```

```python
# Drop rows with any missing values
train_df_cleaned = train_df.dropna().reset_index(drop=True)

# Verify that there are no missing values left
missing_data_cleaned = train_df_cleaned.isnull().sum()
missing_data_cleaned
```

```python
train_df_cleaned.reset_index(drop=True)
```

```python
sample_pages = train_df_cleaned['Page'].sample(10000)
```

```python
df_sample = train_df_cleaned[train_df_cleaned['Page'].isin(sample_pages)]
```

```python
df_sample = df_sample.reset_index(drop=True)
```

```python
df_sample
```

```python
df_sample_cluster = df_sample.copy()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Extract date columns for normalization
date_columns = df_sample_cluster.columns[1:]

# Define a function to apply MinMaxScaler to each row
def scale_row(row):
    scaler = StandardScaler()
    row_scaled = scaler.fit_transform(row.values.reshape(-1, 1)).flatten()
    return pd.Series(row_scaled, index=row.index)

# Apply normalization to each row
df_sample_cluster[date_columns] = df_sample_cluster[date_columns].apply(scale_row, axis=1)
```

```python
df_sample_cluster
```

```python
# Determine the optimal number of clusters using the elbow method
sse = []  # Sum of squared distances to the closest cluster center
from sklearn.metrics import silhouette_score


for num_clusters in range(1, 30,1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans.fit(df_sample_cluster[date_columns])
    sse.append(kmeans.inertia_)  # Inertia: Sum of squared distances of samples to their closest cluster center
```

```python
# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 30,1), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances (Inertia)')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Initialize lists to store SSE and silhouette scores
sse = []
silhouette_scores = []

# Assuming df_sample_cluster contains your data and date_columns contains the columns you want to cluster on

for num_clusters in range(2, 20,1):
    kmeans = KMeans(n_clusters=num_clusters, random_state=0)
    kmeans.fit(df_sample_cluster[date_columns])
    
    # Calculate SSE (Inertia)
    sse.append(kmeans.inertia_)
    
    # Calculate silhouette score
    silhouette_scores.append(silhouette_score(df_sample_cluster[date_columns], kmeans.labels_))

# Plot the elbow graph
plt.figure(figsize=(12, 5))

# Plot SSE
plt.subplot(1, 2, 1)
plt.plot(range(2, 500,20), sse, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Sum of squared distances (Inertia)')
plt.title('Elbow Method for Optimal Number of Clusters')

# Plot silhouette score
plt.subplot(1, 2, 2)
plt.plot(range(2, 500,20), silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal Number of Clusters')

plt.tight_layout()
plt.show()
```
